```python
import numpy as np
import matplotlib.pyplot as plt

#Q1-1
trials = 500
N_values = [1, 10, 500]

bins = 10

for N in N_values:
    sample_means = [0]*trials

    for i in range(trials):
        samples = np.random.choice([0, 1], size=N, p=[0.3, 0.7])
        sample_means[i] = np.mean(samples)

    plt.hist(sample_means, bins = np.linspace(0, 1, 10))
    plt.title(f'Histogram of Sample N = {N}')
    plt.xlabel('Sample Means of X')
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Q1-2


p0 = 0.3
p1 = 0.7


mu = p1
epsilon = 0.1
num_trials = 10


freq =  [0]*100
hoeffding_bounds = [0]*100


for N in range(1, 101):
    hoeffding_bounds[N-1] = 2 * np.exp(-2 * epsilon**2 * N) 

    
for N in range(1, 101):
    num_large_epsilons = 0
    for i in range(num_trials):
        X_samples = np.random.choice([0, 1], size=N, p=[p0, p1])
        X_mean = np.mean(X_samples)
        if abs(X_mean - mu) >= epsilon:
            num_large_epsilons += 1
    freq[N-1] = num_large_epsilons / num_trials
    
    

plt.plot(range(1, 101), freq, label='Empirical')
plt.plot(range(1, 101), hoeffding_bounds, label='Hoeffding Bound')
plt.xlabel('N')
plt.ylabel('P(|X N  |  )')
plt.legend()
plt.show()
```

```python
import math
import random

mu_1 = 0.8
mu_2 = 0.4


# 2.1
def regret_1(T):
    regret = [0]* len(T)
    for t in T:

        N = int(np.ceil(0.3 * t))
        
        
        coin_1 = np.random.binomial(1, 0.8, N)
        coin_2 = np.random.binomial(1, 0.4, N)


        coin_1_empirical = np.sum(coin_1)
        coin_2_empirical = np.sum(coin_2)

        
        if coin_1_empirical >= coin_2_empirical:
            regret[t-10] = max(mu_1, mu_2) * t - mu_2 * N - mu_1 * (t-N) #T sigma(ci),
        else:
            regret[t-10] = max(mu_1, mu_2) * t - mu_1 * N - mu_2 * (t-N)
        
    return regret

# 2.2
def regret_2(T):
    
    
    regret = [0]* len(T)
    for t in T:

        N = int(np.ceil(0.5*pow(t, 2/3)*pow(math.log(t), 1/3)))

        coin_1 = np.random.binomial(1, 0.8, N)
        coin_2 = np.random.binomial(1, 0.4, N)

        
        coin_1_empirical = np.sum(coin_1)
        coin_2_empirical = np.sum(coin_2)
        
        if coin_1_empirical >= coin_2_empirical:

            regret[t-10] = max(mu_1, mu_2) * t - mu_2 * N - mu_1 * (t-N)
        else:

            regret[t-10] = max(mu_1, mu_2) * t - mu_1 * N - mu_2 * (t-N)
        
    return regret

# 2.3
def regret_3(T, eps):

    regret = [0]* len(T)
    for i in T:
        coin_1_n = coin_2_n = coin_1_empirical = coin_2_empirical = cout=0
        for t in range(i):
            if t == 0:
                coin_1_empirical = np.random.binomial(1, 0.8)
                coin_2_empirical = np.random.binomial(1, 0.4)
                coin_1_n += 1
                coin_2_n += 1
            else:
                print
                if np.random.binomial(1, 1 - eps) == 1:       
                    if coin_1_empirical >= coin_2_empirical:
                        coin_1 = np.random.binomial(1, 0.8)
                        coin_1_n += 1
                        coin_1_empirical = coin_1_empirical + (coin_1 - coin_1_empirical) / coin_1_n
                        cout+=0.8
                    else:
                        coin_2 = np.random.binomial(1, 0.4)
                        coin_2_n += 1
                        coin_2_empirical = coin_2_empirical + (coin_2 - coin_2_empirical) / coin_2_n
                        cout += 0.4
                else:
                    
                    if coin_1_empirical >= coin_2_empirical:
                        coin_2 = np.random.binomial(1, 0.4)
                        coin_2_n += 1
                        coin_2_empirical = coin_2_empirical + (coin_2 - coin_2_empirical) / coin_2_n
                        cout += 0.4
                    else:
                        coin_1 = np.random.binomial(1, 0.8)
                        coin_1_n += 1
                        coin_1_empirical = coin_1_empirical + (coin_1 - coin_1_empirical) / coin_1_n
                        cout+=0.8
                        
                        
        regret[i-10]=max(mu_1, mu_2) * t - cout
    return regret


def regret_4(T):
    regret = []
    for i in T:
        coin_1_n = coin_2_n = coin_1_empirical = coin_2_empirical = cout=0
        for t in range(i):
            if t == 0:
                coin_1_empirical = np.random.binomial(1, 0.8)
                coin_2_empirical = np.random.binomial(1, 0.4)
                coin_1_n += 1
                coin_2_n += 1
            else:
                    if coin_1_empirical + math.sqrt(2*math.log(t)/coin_1_n) >= coin_2_empirical + math.sqrt(2*math.log(t)/coin_2_n):
                        coin_1 = np.random.binomial(1, 0.8)
                        coin_1_n += 1
                        coin_1_empirical = coin_1_empirical + (coin_1 - coin_1_empirical) / coin_1_n
                        cout+=0.8
                    else:
                        coin_2 = np.random.binomial(1, 0.4)
                        coin_2_n += 1
                        coin_2_empirical = coin_2_empirical + (coin_2 - coin_2_empirical) / coin_2_n
                        cout += 0.4
        regret.append(max(mu_1, mu_2) * t - cout)
    return regret
```
